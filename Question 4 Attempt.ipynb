```python
# task 3 of 5

# This is the sales data in the question
sales_data = {
    2019: {"Laptops": 225, "Desktops": 0, "Phones": 0, "Tablets": 0},
    2020: {"Laptops": 385, "Desktops": 945, "Phones": 854, "Tablets": 0},
    2021: {"Laptops": 359, "Desktops": 252, "Phones": 1249, "Tablets": 108},
    2022: {"Laptops": 457, "Desktops": 287, "Phones": 1556, "Tablets": 219},
    2023: {"Laptops": 429, "Desktops": 124, "Phones": 1609, "Tablets": 283},
    2024: {"Laptops": 342, "Desktops": 0, "Phones": 1973, "Tablets": 356},
}

# This Function should find the highest selling year for each device
def highest_selling_year(sales_data):
    highest_sales = {
        "Laptops": (0, 0),
        "Desktops": (0, 0),
        "Phones": (0, 0),
        "Tablets": (0, 0),
    }
    
    for year, products in sales_data.items():
        for product, sales in products.items():
            if sales > highest_sales[product][1]:
                highest_sales[product] = (year, sales)
    
    return highest_sales

highest_sales = highest_selling_year(sales_data)

for product, (year, sales) in highest_sales.items():
    print(f"Highest selling year for {product} was {year} with {sales} units sold.")
```

```python
# Task 4 of 5
# Going to use matplotlib for this question, which is a plotting library. Installed in my envroment using 'pip install matplotlib' 

import matplotlib.pyplot as plt

# The sales data from orignal question
sales_data = {
    2019: {"Laptops": 225, "Desktops": 0, "Phones": 0, "Tablets": 0},
    2020: {"Laptops": 385, "Desktops": 945, "Phones": 854, "Tablets": 0},
    2021: {"Laptops": 359, "Desktops": 252, "Phones": 1249, "Tablets": 108},
    2022: {"Laptops": 457, "Desktops": 287, "Phones": 1556, "Tablets": 219},
    2023: {"Laptops": 429, "Desktops": 124, "Phones": 1609, "Tablets": 283},
    2024: {"Laptops": 342, "Desktops": 0, "Phones": 1973, "Tablets": 356},
}

# Extract years/laptop sales
years = list(sales_data.keys())
laptop_sales = [sales_data[year]["Laptops"] for year in years]

# Plotting data
plt.figure(figsize=(10, 5))
plt.plot(years, laptop_sales, marker='o', linestyle='-', color='b', label='Laptops')

# Add the labels and title
plt.xlabel('Year')
plt.ylabel('# of Laptops Sold')
plt.title('Laptop Sales Over Years')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Task 5 of 5

# Sales data
sales_data = {
    2019: {"Laptops": 225, "Desktops": 0, "Phones": 0, "Tablets": 0},
    2020: {"Laptops": 385, "Desktops": 945, "Phones": 854, "Tablets": 0},
    2021: {"Laptops": 359, "Desktops": 252, "Phones": 1249, "Tablets": 108},
    2022: {"Laptops": 457, "Desktops": 287, "Phones": 1556, "Tablets": 219},
    2023: {"Laptops": 429, "Desktops": 124, "Phones": 1609, "Tablets": 283},
    2024: {"Laptops": 342, "Desktops": 0, "Phones": 1973, "Tablets": 356},
}

def get_sales_data():
    while True:
        product = input("Enter the product name (Laptops, Desktops, Phones, Tablets): ").capitalize()
        year = input("Enter Year")

        if not year.isdigit():
            print("Invalid year. Enter a valid year.")
            continue

        year = int(year)

        if year not in sales_data:
            print("Year not found. Please enter a valid year.")
            continue

        if product not in sales_data[year]:
            print("Product not found. Please enter a valid product.")
            continue

        sales = sales_data[year][product]
        print(f"The sales for {product} in {year} were {sales} units.")
        break

# Call the function to prompt the user
get_sales_data()
```
